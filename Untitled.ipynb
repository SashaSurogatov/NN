```python
#import os
#os.environ["THEANO_FLAGS"] = "mode=FAST_RUN,device=gpu,floatX=float32"
from __future__ import print_function
from scipy import misc
import os
import numpy as np
import keras
import h5py as h5py
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D

batch_size = 64
num_classes = 2
epochs = 12

# input image dimensions
img_rows, img_cols = 256, 256

model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(img_rows, img_cols, 1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

x = []
y = []
xt = []
yt = []
        
sick_path = '../../Data/Sick/Normalized/'
healthy_path = '../../Data/'

sick_map = {}
with open(sick_path + 'idx.txt', 'r') as f:
    for line in f:
        path, age, sex = line.split(',')
        if (age, sex) in sick_map:
            sick_map[(age, sex)].append(path)
        else:
            sick_map[(age, sex)] = [path]
      
print (len(sick_map))
for key, value in sick_map.items():
    
    if int(key[0]) < 21 or int(key[0]) > 70:
        continue
        
    l = int(0.7 * len(value))
    files = os.listdir(healthy_path + str(key[0]) + '_' + key[1][0] + '/')
    
    for i in range(0, l):
        image = misc.imread(sick_path + value[i])
        x.append(image)
        y.append(1)  

        image = misc.imread(healthy_path + str(key[0]) + '_' + key[1][0] + '/' + files[i])
        x.append(image)
        y.append(0) 
    for i in range(l, len(value)):
        image = misc.imread(sick_path + value[i])
        xt.append(image)
        yt.append(1)

        mage = misc.imread(healthy_path + str(key[0]) + '_' + key[1][0] + '/' + files[i])
        xt.append(image)
        yt.append(0)         

x_train = np.array(x)
y_train = np.array(y)
x_test = np.array(xt)
y_test = np.array(yt)

            
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)

x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

print('y_train shape:', y_train.shape)
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print (y_train)
print (y_test)

model.fit(x_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            verbose=1,
            validation_data=(x_test, y_test))

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
    
model.save('lenet_model.m5')
```

```python
import numpy as np
import keras
from keras.models import load_model
from scipy import misc

img_rows, img_cols = 256, 256
num_classes = 2

model = load_model('lenet_model.m5')

path = '../../Data/train-val-splits/2-ages_junior-senior_5-fold/'

val_file_name = 'idx_labeled_val_fold-%d.txt' % (4)

f_val = open(path + val_file_name, 'r')

xt = []
yt = []

for line in f_val.readlines():
    file_name, class_val = line.split()
    image = misc.imread('../../Data/' + file_name)
    xt.append(image)
    yt.append(class_val)
    
x_test = np.array(xt)
y_test = np.array(yt)

x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
y_test = keras.utils.to_categorical(y_test, num_classes)

predicted = model.predict(x_test);

correct = 0
for i in range(0, predicted.shape[0]):
    #print ('X = ', predicted[i,0:], '\nY = ', y_test[i,0:])
    diff = predicted[i,0:] - y_test[i,0:]
    if (diff[0] < 1/1e5 and diff[1] < 1/1e5):
        correct += 1

print (predicted.shape[0], correct, correct * 100 / predicted.shape[0])
```
